```python
import pandas as pd 

df = pd.read_parquet('data/dados_clones.parquet')

df
```

```python
df.columns
```

```python
variable = ['Massa(em kilos)', 'Estatura(cm)' ]

df.groupby('Status ')[variable].mean()
```

```python
X = pd.get_dummies(df.drop(columns=['p2o_master_id', 'Status ', 'General Jedi encarregado' ]))
y = df['Status ']
```

```python
from sklearn import tree

model = tree.DecisionTreeClassifier()

model.fit(X, y)
```

```python
import matplotlib.pyplot as plt 

plt.figure(dpi=400)

tree.plot_tree(model, feature_names=X.columns, 
               class_names=model.classes_,
               max_depth=3, 
               rounded=True,
               filled=True)
```
