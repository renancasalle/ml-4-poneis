```python
import pandas as pd

df = pd.read_excel('data/dados_cerveja.xlsx')

df
```

```python
X = pd.get_dummies(df.drop(columns=['classe', 'id']))
y = df['classe']
```

```python
from sklearn import tree

model = tree.DecisionTreeClassifier()

model.fit(X, y)
```

```python
import matplotlib.pyplot as plt

plt.figure(dpi=400)

tree.plot_tree(model, feature_names=X.columns, 
               class_names=model.classes_, 
               filled=True)
```
